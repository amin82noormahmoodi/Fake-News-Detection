```python
def is_suspecious(authors, author_name):
    log = authors[authors['author'] == author_name]
    news_count = log['label'].value_counts()
    try:
        reals = news_count['Real']
    except:
        reals = 0
    try:
        fakes = news_count['Fake']
    except:
        fakes = 0
        
    if fakes > reals:
        return 1
    
    return -1 
```

```python
authors = train[["author", "label"]]
authors
```

```python
train["is suspecious"] = train["author"].map(lambda x:is_suspecious(authors, x))
```
